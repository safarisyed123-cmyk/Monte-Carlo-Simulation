```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

# Basic inputs
initial_price = 1000
expected_annual_return = 0.08
annual_volatility = 0.20
years = 7
trading_days = 252
simulations = 15000

# Convert years to total days
total_days = years * trading_days

# Convert annual values to daily values
daily_return = expected_annual_return / trading_days
daily_volatility = annual_volatility / np.sqrt(trading_days)

# Generate random daily returns
returns_array = np.random.normal(
    daily_return,
    daily_volatility,
    (total_days, simulations)
)

# Calculate price paths
price_paths = initial_price * np.cumprod(1 + returns_array, axis=0)

# Final prices after 7 years
final_prices = price_paths[-1]

# Plot histogram
plt.hist(final_prices, bins=10)
plt.title("Monte Carlo Simulation - 7 Years")
plt.xlabel("Final Price")
plt.ylabel("Frequency")
plt.show()

# Probability example
probability_above_1500 = np.mean(final_prices > 1500)

print("Probability price above 1500 after 7 years:")
print(probability_above_1500)
print(len(final_prices))
```
